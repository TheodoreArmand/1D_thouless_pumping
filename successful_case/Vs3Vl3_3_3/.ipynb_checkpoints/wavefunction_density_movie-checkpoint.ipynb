# Vs3/Vl3 3/3 wavefunction density movie

这个 notebook 读取成功案例的 `snapshots.csv`，把每个 snapshot 保存的 ECG 基函数参数重建成一维波函数密度。它的目标是方便拖进度条看 density 演化，并导出 `density_evolution.mp4` 或 `density_evolution.gif`。物理量判断仍以 `trace.csv` 和 `summary.txt` 为准。

```python
from pathlib import Path
import os

os.environ.setdefault('MPLCONFIGDIR', '/tmp/matplotlib-ecg1d')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML, Video, display

def find_project_root(start=Path.cwd()):
    run_rel = Path('out/pump_vs3pad_gapadaptive_T160pi/a8p000_K16_tmax502p655_VsER3p000_VlER3p000')
    for path in [start, *start.parents]:
        if (path / run_rel / 'snapshots.csv').exists():
            return path
    raise FileNotFoundError('找不到 3/3 成功案例 snapshots.csv；请从 1D_thouless_pumping 或本 notebook 目录启动 Jupyter。')

PROJECT_ROOT = find_project_root()
RUN_DIR = PROJECT_ROOT / 'out/pump_vs3pad_gapadaptive_T160pi/a8p000_K16_tmax502p655_VsER3p000_VlER3p000'
SNAPSHOT_CSV = RUN_DIR / 'snapshots.csv'
TRACE_CSV = RUN_DIR / 'trace.csv'
OUT_DIR = PROJECT_ROOT / 'successful_case/Vs3Vl3_3_3'
OUT_DIR.mkdir(parents=True, exist_ok=True)

print('PROJECT_ROOT =', PROJECT_ROOT)
print('SNAPSHOT_CSV =', SNAPSHOT_CSV)
print('TRACE_CSV =', TRACE_CSV)
```

```python
snapshots = pd.read_csv(SNAPSHOT_CSV)
trace = pd.read_csv(TRACE_CSV)

snapshot_groups = [(int(sid), frame.reset_index(drop=True)) for sid, frame in snapshots.groupby('snapshot', sort=True)]
print(f'snapshot groups: {len(snapshot_groups)}')
print(f'rows: {len(snapshots)}')
display(snapshots.head())
display(trace.head())
```

## ECG density reconstruction

对当前 N=1 数据，每一行 snapshot 对应一个 Gaussian basis。按照 `src/pair_cache.cpp` 里的积分约定，单个基函数按下面形式可视化：

$$g_k(x)=\exp\left[-(A_k+B_k)x^2+2B_kR_kx-B_kR_k^2\right].$$

总波函数用 $\psi(x)=\sum_k u_k g_k(x)$，density 用 $|\psi(x)|^2$，每一帧单独归一化后画图。

```python
def make_x_grid(trace_df, n=1600, sigma_factor=4.0, extra=2.0):
    x_mean = trace_df['x_mean'].to_numpy(float)
    x2 = trace_df['x2'].to_numpy(float)
    variance = np.maximum(x2 - x_mean * x_mean, 0.0)
    sigma = np.sqrt(variance)
    lo = np.nanmin(x_mean - sigma_factor * sigma) - extra
    hi = np.nanmax(x_mean + sigma_factor * sigma) + extra
    return np.linspace(lo, hi, n)

x = make_x_grid(trace)

def frame_wavefunction(frame, x_grid):
    u = frame['u_re'].to_numpy(float) + 1j * frame['u_im'].to_numpy(float)
    A = frame['A_re'].to_numpy(float) + 1j * frame['A_im'].to_numpy(float)
    B = frame['B_re'].to_numpy(float) + 1j * frame['B_im'].to_numpy(float)
    R = frame['R_re'].to_numpy(float) + 1j * frame['R_im'].to_numpy(float)
    xx = x_grid[None, :]
    exponent = -(A + B)[:, None] * xx * xx + (2.0 * B * R)[:, None] * xx - (B * R * R)[:, None]
    # 只为避免 exp overflow；density 会重新归一化，所以公共缩放不影响形状。
    shift = np.nanmax(np.real(exponent))
    return (u[:, None] * np.exp(exponent - shift)).sum(axis=0)

def frame_density(frame, x_grid):
    psi = frame_wavefunction(frame, x_grid)
    rho = np.abs(psi) ** 2
    area = np.trapz(rho, x_grid)
    if np.isfinite(area) and area > 0:
        rho = rho / area
    return rho

def trace_row_for_frame(frame):
    t = float(frame['t'].iloc[0])
    idx = int(np.argmin(np.abs(trace['t'].to_numpy(float) - t)))
    return trace.iloc[idx]
```

```python
def plot_density(frame_index=0):
    sid, frame = snapshot_groups[int(frame_index)]
    rho = frame_density(frame, x)
    tr = trace_row_for_frame(frame)

    fig, ax = plt.subplots(figsize=(9, 4.8))
    ax.plot(x, rho, lw=1.8)
    ax.axvline(float(tr['x_mean']), color='black', lw=1.0, alpha=0.45)
    ax.set_xlabel('x')
    ax.set_ylabel('normalized density')
    ax.set_title(f"snapshot={sid}, t={float(frame['t'].iloc[0]):.6g}, phi={float(frame['phi'].iloc[0]):.6g}, polarization={float(tr['polarization_cell']):.6g}")
    ax.set_xlim(x[0], x[-1])
    ax.set_ylim(bottom=0)
    fig.tight_layout()
    return fig

try:
    import ipywidgets as widgets
    widgets.interact(
        plot_density,
        frame_index=widgets.IntSlider(value=0, min=0, max=len(snapshot_groups) - 1, step=1, continuous_update=False),
    )
except ImportError:
    plot_density(0)
```

```python
# Movie settings. 10055 saved snapshots 全部做视频太长；默认每 20 个 saved snapshot 取一帧。
FRAME_STRIDE = 20
FPS = 20
movie_indices = np.arange(0, len(snapshot_groups), FRAME_STRIDE, dtype=int)

fig, ax = plt.subplots(figsize=(9, 4.8))
line, = ax.plot([], [], lw=1.8)
marker = ax.axvline(0.0, color='black', lw=1.0, alpha=0.45)
label = ax.text(0.02, 0.96, '', transform=ax.transAxes, va='top', ha='left')
ax.set_xlim(x[0], x[-1])
movie_ymax = max(frame_density(snapshot_groups[i][1], x).max() for i in movie_indices)
ax.set_ylim(0, 1.05 * movie_ymax)
ax.set_xlabel('x')
ax.set_ylabel('normalized density')

def update_movie(j):
    sid, frame = snapshot_groups[int(movie_indices[j])]
    rho = frame_density(frame, x)
    tr = trace_row_for_frame(frame)
    line.set_data(x, rho)
    marker.set_xdata([float(tr['x_mean']), float(tr['x_mean'])])
    label.set_text(f"snapshot={sid}\nt={float(frame['t'].iloc[0]):.3f}, phi={float(frame['phi'].iloc[0]):.3f}\nP={float(tr['polarization_cell']):.6f}")
    return line, marker, label

anim = FuncAnimation(fig, update_movie, frames=len(movie_indices), interval=1000 / FPS, blit=True)
plt.close(fig)

# 在 notebook 里预览。帧数较多时 to_jshtml 会生成较大的 HTML，可以只保存视频。
# display(HTML(anim.to_jshtml()))

mp4_path = OUT_DIR / 'density_evolution.mp4'
gif_path = OUT_DIR / 'density_evolution.gif'
try:
    anim.save(mp4_path, writer='ffmpeg', fps=FPS, dpi=140)
    print('wrote', mp4_path)
    display(Video(str(mp4_path), embed=False))
except Exception as exc:
    print('mp4 failed, falling back to gif:', exc)
    anim.save(gif_path, writer='pillow', fps=FPS)
    print('wrote', gif_path)
```
